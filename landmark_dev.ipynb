# How to write a function under a module?

Here we give an example of creating a function under the `extraction` module. The process is the same for `derivation`.

- **Step 1: Create a `.py` file**

  For example, we create a file called `load_mask.py`.

- **Step 2: Define the function inside `load_mask.py`**

  Open `load_mask.py`, and write your function like the following. Please only import necessary libraries. Import redundant libraries will make the initilization slower.

  ```python
  import numpy as np

  def load_mask(mask_path):
      """
      Load a binary mask from the given path.

      Args:
          mask_path (str): Path to the mask file.

      Returns:
          np.ndarray: Binary mask as a NumPy array.
      """
      mask = np.load(mask_path)
      return mask
  ```

- **Step 3: Import your function in `extraction/__init__.py`**

  Open the file `__init__.py` under `extraction/` directory. Then add the following code to import the function you just wrote.

  ```python
  from .load_mask import load_mask
  ```

# What if a `.py` file contains multiple functions?

This case may happen if you want to have a `utils.py` file, where you can put your helper functions inside. Here we give an example of creating utility functions directly under the `landmark` module.

- **Step 1: Create a `.py` file**

  For example, we create a file called `utils.py`.

- **Step 2: Define the functions inside `utils.py`**

  Open `utils.py`, and write your functions like the following. Please only import necessary libraries. Import redundant libraries will make the initilization slower.

  ```python
  import math
  import json

  def calculate_distance(p1: tuple[float, float], p2: tuple[float, float]) -> float:
      """
      Calculate the Euclidean distance between two 2D points.

      Args:
          p1 (tuple): Coordinates of the first point (x1, y1).
          p2 (tuple): Coordinates of the second point (x2, y2).

      Returns:
          float: Euclidean distance between the two points.
      """
      return math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)

  def load_json(file_path: str) -> dict:
      """
      Load a JSON file into a Python dictionary.

      Args:
          file_path (str): Path to the JSON file.

      Returns:
          dict: Parsed JSON content.
      """
      with open(file_path, 'r', encoding='utf-8') as f:
          return json.load(f)
  ```

# How to use functions in `utils.py` in other Python functions

Please see the following code as an example. Suppose you want to compare if the first distance is larger than the second one, only using coordinates.

````python
# Import like this
from garmentiq.landmark.utils import calculate_distance
from typing import Tuple

def compare_distance(
    coord_set_1: Tuple[Tuple[float, float], Tuple[float, float]],
    coord_set_2: Tuple[Tuple[float, float], Tuple[float, float]],
) -> None:
    """
    Compare the distances between two sets of coordinate pairs
    and print whether the first distance is greater than the second.

    Args:
        coord_set_1: A tuple of two (x, y) coordinate pairs.
        coord_set_2: A tuple of two (x, y) coordinate pairs.
    """
    distance_1 = calculate_distance(coord_set_1[0], coord_set_1[1])
    distance_2 = calculate_distance(coord_set_2[0], coord_set_2[1])

    print(distance_1 > distance_2)
````

````python
# Usage
# Define two sets of point pairs
coord_set_1 = ((1.0, 2.0), (4.0, 6.0))   # distance в‰€ 5.0
coord_set_2 = ((2.0, 3.0), (3.0, 4.0))   # distance в‰€ 1.41

# Compare and print whether distance_1 > distance_2
compare_distance(coord_set_1, coord_set_2)
````
